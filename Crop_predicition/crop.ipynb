```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler   # it use to normalise the outliers

std = StandardScaler()
x_transformed = std.fit_transform(x)
```

```python
x_transformed
```

```python
print('total number of clay samples :',x_transformed.shape[0])
```

```python
# traning and testing splitig
train = x_transformed[:2100]
test = x_transformed[2100:]
```

```python
# this is unsupervised machine learning 
from sklearn.cluster import KMeans
obj = KMeans()
```

```python
# fitted on tranin data 
obj.fit(train)
```

```python
clust = obj.labels_
```

```python
testing_df = x[:2100]
```

```python
testing_df['cluster'] = clust
```

```python
testing_df['label'] = df[:2100]['label']
```

```python
testing_df['cluster'].value_counts()
```

```python
testing_df
```

```python
fifth_cluster = testing_df[testing_df['cluster']==5]
```

```python
print('Total points in the 5th cluster',fifth_cluster.shape[0])
```

```python
list(fifth_cluster['label'].unique())
```

```python
obj.score(train,clust)
```

```python
# to calculate the score 
# scoreing is calclute -1 t0 +1

from sklearn.metrics import silhouette_score
```

```python
print('Your score is :',silhouette_score(train,clust)*100)
```
